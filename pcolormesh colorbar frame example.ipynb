```python
## Goal: Use pcolormesh with a colorbar, but the colorbar shouldn't have a frame.

import matplotlib.pyplot as plt 
import numpy as np

Z = np.random.rand(5, 5)

plt.pcolormesh(Z)
bigax = plt.gca()
bigax.set_frame_on(False)
cbar = plt.colorbar(drawedges=False)
cbar.ax.set_frame_on(False) # this line doesn't work
cbar.outline.set_visible(False) 

# For contrast, seaborn's heatmap has a colorbar without a frame. Can I do it with pcolormesh??
import seaborn as sns
plt.figure()
sns.heatmap(Z, cmap = 'viridis')


plt.show()
```
